### 1. ****IMPORT REQUIRED LIBRARIES****


```python

# OS -> folder/file handling
import os

# For visualization
import matplotlib.pyplot as plt

# For image processing
import cv2
import random

print("Libraries imported successfully!")
```

### 2. ****DEFINE DATASET PATH****

```python

# Path where raw dataset is stored
DATA_DIR = "../data/raw"

# List all class folders
classes = os.listdir(DATA_DIR)

print("Classes found in dataset:")
print(classes)
print("\nTotal number of classes:", len(classes))
```

### 3. ****COUNT IMAGES IN EACH CLASS****

```python


class_counts = []

for cls in classes:
    class_path = os.path.join(DATA_DIR, cls)
    num_images = len(os.listdir(class_path))
    class_counts.append(num_images)

# Print count for each class
for i in range(len(classes)):
    print(f"Class: {classes[i]} --> Images: {class_counts[i]}")
```

### 4. ****VISUALIZE CLASS DISTRIBUTION****

```python


plt.figure(figsize=(10,5))
plt.bar(classes, class_counts)
plt.title("Class Distribution (Number of Images per Class)")
plt.xlabel("Classes")
plt.ylabel("Number of Images")
plt.xticks(rotation=45)
plt.show()
```

### 5. ****DISPLAY SAMPLE IMAGES FROM DATASET****

```python

plt.figure(figsize=(10,10))

for i in range(9):
    
    # Randomly select a class
    cls = random.choice(classes)
    
    # Randomly select image from that class
    img_name = random.choice(os.listdir(os.path.join(DATA_DIR, cls)))
    
    img_path = os.path.join(DATA_DIR, cls, img_name)
    
    # Read image
    img = cv2.imread(img_path)
    
    # Convert BGR to RGB (OpenCV loads in BGR)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Plot image
    plt.subplot(3,3,i+1)
    plt.imshow(img)
    plt.title(cls)
    plt.axis('off')

plt.show()
```

### 6. ****CHECK IMAGE SHAPE****

```python


sample_class = classes[0]
sample_img = os.listdir(os.path.join(DATA_DIR, sample_class))[0]

img_path = os.path.join(DATA_DIR, sample_class, sample_img)

img = cv2.imread(img_path)

print("Original Image Shape:", img.shape)
```
